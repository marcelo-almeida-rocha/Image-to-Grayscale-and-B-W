**Imagem colorida para Tons de Cinza e Preto e Branco**

```python
from google.colab import files
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

Função para converter para tons de cinza

```python
def to_grayscale(image):
    width, height = image.size
    grayscale_image = Image.new('L', (width, height))  # 'L' para tons de cinza
    pixels = image.load()
    gray_pixels = grayscale_image.load()

    for x in range(width):
        for y in range(height):
            r, g, b = pixels[x, y]
            # Fórmula de luminância perceptual
            gray = int(0.2989 * r + 0.5870 * g + 0.1140 * b)
            gray_pixels[x, y] = gray

    return grayscale_image
```

Função para converter para preto e branco (binário)

```python
def to_binary(grayscale_image, threshold=127):
    width, height = grayscale_image.size
    binary_image = Image.new('1', (width, height))  # '1' para imagens binárias
    gray_pixels = grayscale_image.load()
    binary_pixels = binary_image.load()

    for x in range(width):
        for y in range(height):
            # Se o pixel for maior que o threshold, torna branco, caso contrário, preto
            binary_pixels[x, y] = 1 if gray_pixels[x, y] > threshold else 0

    return binary_image
```

```python
# Upload da imagem
print("Faça o upload da imagem:")
uploaded = files.upload()
```

```python
# Processar a imagem
image_path = next(iter(uploaded.keys()))
image = Image.open(image_path).convert('RGB')  # Converte para RGB se não estiver
```

```python
# Converter para tons de cinza
grayscale_image = to_grayscale(image)
```

```python
# Converter para preto e branco
binary_image = to_binary(grayscale_image)
```

```python
# Exibir as imagens
plt.figure(figsize=(15, 5))
```

```python
# Imagem original
plt.subplot(1, 3, 1)
plt.title("Imagem Colorida")
plt.imshow(image)
plt.axis('off')

# Imagem em tons de cinza
plt.subplot(1, 3, 2)
plt.title("Tons de Cinza")
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')

# Imagem binária
plt.subplot(1, 3, 3)
plt.title("Preto e Branco (Binária)")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')

plt.show()
```
